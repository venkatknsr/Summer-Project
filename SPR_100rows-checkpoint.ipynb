```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Scale target variable
y_scaler = MinMaxScaler()
y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1, 1))

# Build the model
model = Sequential([
    Dense(units=128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dense(units=64, activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=1)
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='mean_squared_error')

# Implement early stopping
#early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train the model
history = model.fit(
    X_train_scaled,
    y_train_scaled,
    epochs=20,
    batch_size=32,
    validation_split=0.2,
)

# Evaluate the model
loss = model.evaluate(X_test_scaled, y_test_scaled)
print(f'Test Loss: {loss}')

# Make predictions
predictions_scaled = model.predict(X_test_scaled)

# Inverse transform predictions
predictions = y_scaler.inverse_transform(predictions_scaled)

# Print or analyze predictions
print(predictions)
r21=r2_score(predictions,y_test)
```

```python
# Calculate MAE, MSE, RMSE, and RÂ² for the test set
y_test_pred = model.predict(X_test_scaled)

# Inverse transform predictions and true values
y_test = y_test.values.reshape(-1, 1)
y_test_inv = y_scaler.inverse_transform(y_test_scaled)
y_test_pred_inv = y_scaler.inverse_transform(y_test_pred)

# Calculate metrics
mae = mean_absolute_error(y_test_inv, y_test_pred_inv)
mse = mean_squared_error(y_test_inv, y_test_pred_inv)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_inv, y_test_pred_inv)

# Print metrics
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (RÂ²): {r2}')
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_csv('ml_new_dataset.csv')
data=data.sort_values(by='Locality').reset_index()
data.head()
data.drop(columns=['index'],inplace=True)
data.head()
```

```python
data['Balcony'].fillna(0,inplace=True)
data['Parking'].fillna('No',inplace=True)
```

```python
en=pd.get_dummies(data[['Locality','Furniture','Age','Parking','Preferred Tenant']]).astype(int)
```

```python
data.drop(columns=['Locality','Furniture','Age','Parking','Preferred Tenant'],inplace=True)
data.head()
```

```python
x=data.iloc[:,0:5]
x
```

```python
df=pd.concat([x,en],axis=1)
```

```python
y=df['Rent Price']
df.drop(columns='Rent Price',axis=1)
```

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(df, y, test_size=0.2, random_state=42)
X_train,X_val,y_train,y_val=train_test_split(X_train_val,y_train_val,test_size=0.25,random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_val_scaled=scaler.transform(X_val)
```

```python
from sklearn.tree import DecisionTreeRegressor
model_s=DecisionTreeRegressor(random_state=42)
model_s.fit(X_train_scaled,y_train)
y_pred=model_s.predict(X_test_scaled)
from sklearn.metrics import r2_score
r2_s = r2_score(y_test, y_pred)
print(r2_s)
```

```python
train_predictions = model_s.predict(X_train_scaled)
train_mse = mean_squared_error(y_train, train_predictions)
train_rmse = np.sqrt(train_mse)
print("Training RMSE:", train_rmse)
```

```python
val_predictions = model_s.predict(X_val_scaled)
val_mse = mean_squared_error(y_val, val_predictions)
val_rmse = np.sqrt(val_mse)
print("Validation RMSE:", val_rmse)
```

```python
test_predictions = model_s.predict(X_test_scaled)
test_mse = mean_squared_error(y_test, test_predictions)
test_rmse = np.sqrt(test_mse)
print("Test RMSE:", test_rmse)
#as the training MSE is much lower than val and test MSE the model is overfitting
```
